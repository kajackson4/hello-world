Exercise 6

Rewrite your pay computation with time-and-a-half for overtime (for the purposes of this exercise, let's say that a normal working week is 40 hours) and create a function called `computepay` which takes two parameters (hours and rate).

Example Output:

`Enter Hours: 45`

`Enter Rate: 10`

`Pay: 475.0`

```python
def computepay():
    hours = input('Enter hours worked')
    rate = input('Enter rate per hour')
    overHours = int(hours) / 2
    overRate = int(rate) / 2 
    pay = int(hours) * int(rate) + int(overHours) * int(overRate)
    return pay
```

| Percentage | Grade | 
|------------|-------|
| >= 0.9     | A     |  
| >= 0.8     | B     |   
| >= 0.7     | C     |  
| >= 0.6     | D     |  
| < 0.6      | F     |  

Example Output:

`Enter score: 0.95`

`A`

```python
def computegrade():
    score = input('Enter grade percentage.')
    if int(score) >= 0.9 
    return 'A'
    if int(score) >= 0.8
    return 'B'
    if int(score) >= 0.7
    return 'C'
    if int(score) >= 0.6
    return 'D'
    if int(score) < 0.6
    return 'F'
```
